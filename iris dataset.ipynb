```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("iris_csv.csv")
plt.figure(figsize=(10,7))
x=data["sepallength"]
plt.hist(x,bins=20,color="green")
plt.title("SEPAL LENGTH in cm")
plt.xlabel("sepallength")
plt.ylabel("count")
plt.show()
```
